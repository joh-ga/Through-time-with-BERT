```python
import numpy as np
from sklearn.mixture import GaussianMixture
```

### Clustering Sentences without Adverbs

```python
with open('../data/tensors.tsv') as f:
    tsv = f.readlines()
tensors = [i.split('\t') for i in tsv]
tensorsNum = np.array(tensors, dtype='float32')
tensorsNum

data=np.genfromtxt('../data/tensors.tsv', delimiter='\t', invalid_raise = False)
vecs  = data.astype('float32')
```

```python
with open('../data/metadata.tsv') as f:
    tsv_metadata = f.readlines()
    
metadata = [i.strip().split('\t') for i in tsv_metadata]
metadata.remove(['Id', 'Tense', 'Sentence'])
metadata
```

#### 3 Cluster

```python
gmm = GaussianMixture(n_components=3, covariance_type='diag', max_iter=500, random_state=0).fit(tensorsNum)
#covariance_type='diag': contour axes are oriented along coordinate axes, but otherwise eccentricities ('Abweichungen') may vary between components
print('Converged:', gmm.converged_) #check if the model has converged
prediction = gmm.predict(tensorsNum) #Labels
probs = gmm.predict_proba(tensorsNum) #probability for clusters: output depends on covariance_type
#prediction
probs

###
#OPTIONAL: only 'probs'
##Add cluster labels to assigned probabilities when writing probabilities
cluster_probs=[]
for p in probs:
    i=0
    items=[]
    for j in p:
        items.append((i, j))
        i+=1
    cluster_probs.append(items)
cluster_probs
###
```

```python
cluster_3 = []
i = 0

for c in cluster_probs:
    cluster_3.append((metadata[i][0], c, metadata[i][1], metadata[i][2]))
    i += 1
cluster_3
```

```python
#3 Cluster
#f = open('../data/fuzzyClusterGMM_3.tsv','w') #for labels ('prediction')
f = open('../data/probs_fuzzyClusterGMM_3.tsv','w') #for probabilities ('probs')

#f.write("Index\tCluster\tTense\tSentence\n") #for labels ('prediction')
f.write("Index\t(Cluster, Probability)\tTense\tSentence\n") #for probabilities ('probs')

for i in range(len(cluster_3)):
    f.write(str(cluster_3[i][0])+"\t"+str(cluster_3[i][1])+"\t"+str(cluster_3[i][2])+"\t"+str(cluster_3[i][3])+"\n")
f.close()
```

### Clustering Sentences with Adverbs

```python
with open('../data/tensorsAdverb.tsv') as f:
    tsv = f.readlines()
tensorsAdv = [i.split('\t') for i in tsv]
tensorsAdvNum = np.array(tensorsAdv, dtype='float32')
tensorsAdvNum
```

```python
with open('../data/metadataAdverb.tsv') as f:
    tsv_metadataAdv = f.readlines()
    
metadataAdv = [i.strip().split('\t') for i in tsv_metadataAdv]
metadataAdv.remove(['Id', 'Tense', 'Sentence'])
```

#### 3 Cluster

```python
gmm = GaussianMixture(n_components=3, covariance_type='diag', max_iter=500, random_state=0).fit(tensorsAdvNum)
#covariance_type='diag': contour axes are oriented along coordinate axes, but otherwise eccentricities ('Abweichungen') may vary between components
print('Converged:', gmm.converged_) #check if the model has converged
prediction = gmm.predict(tensorsAdvNum) #Labels
probs = gmm.predict_proba(tensorsAdvNum) #probability for clusters: output depends on covariance_type
score = gmm.score_samples(tensorsNum) #Log-likelihood of each sample
#prediction
probs 

###
#OPTIONAL: only 'probs'
##Add cluster labels to assigned probabilities when writing probabilities
cluster_probs=[]
for p in probs:
    i=0
    items=[]
    for j in p:
        items.append((i, j))
        i+=1
    cluster_probs.append(items)
cluster_probs
###
```

```python
clusterAdv_3 = []
i = 0

#for c in prediction:
for c in cluster_probs:
    clusterAdv_3.append((metadataAdv[i][0], c, metadataAdv[i][1], metadataAdv[i][2]))
    i += 1
clusterAdv_3
```

```python
#3 Cluster
#f = open('../data/fuzzyClusterGMM_Adv_3.tsv','w') #for labels ('prediction')
f = open('../data/probs_fuzzyClusterGMM_Adv_3.tsv','w') #for probabilities ('probs')

#f.write("Index\tCluster\tTense\tSentence\n") #for labels ('prediction')
f.write("Index\t(Cluster, Probability)\tTense\tSentence\n") #for probabilities ('probs')

for i in range(len(clusterAdv_3)):
    f.write(str(clusterAdv_3[i][0])+"\t"+str(clusterAdv_3[i][1])+"\t"+str(clusterAdv_3[i][2])+"\t"+str(clusterAdv_3[i][3])+"\n")
f.close()
```
